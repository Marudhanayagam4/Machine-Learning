```python
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset
```

```python

dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python

dependent = dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, random_state=0, test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
scx = StandardScaler()
x_train = scx.fit_transform(x_train)
x_test = scx.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid = {
    "solver": ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    "penalty": ['l2'],
    "multi_class": ['auto', 'ovr', 'multinomial']
}

grid = GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3, cv=5, n_jobs=-1, scoring="f1_weighted")

grid.fit(x_train, y_train)
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, y_pred, average="weighted")
print("The f1 macro value for the best para metes is {}:".format(grid.best_params_),f1_macro)

from sklearn.metrics import roc_auc_score
roc_score = roc_auc_score(y_test, grid.predict_proba(x_test)[:,1])
print("The roc_auc score for the model is {}:".format(roc_score))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
cm = confusion_matrix(y_test, y_pred)
clf_report = classification_report(y_test, y_pred)

print(cm)
print(clf_report)
```

```python
results = grid.cv_results_
table = pd.DataFrame.from_dict(results)
table
```
