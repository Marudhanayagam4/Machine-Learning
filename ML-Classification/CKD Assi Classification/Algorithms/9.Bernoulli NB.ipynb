```python
import pandas as pd
```

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
independent = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent = dataset[['classification_yes']]
dependentdependent = dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
X_train
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import GridSearchCV
param_grid = {
    'alpha': [0.0, 0.1, 0.5, 1.0, 2.0],
    'binarize': [0.0, 0.5, 1.0],
    'fit_prior': [True, False]
}

grid=GridSearchCV(BernoulliNB(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
y_pred = grid.predict(independent)
y_pred
```

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(dependent, y_pred, average="weighted")
print("The f1 macro value for the best para metes is {}:".format(grid.best_params_),f1_macro)

from sklearn.metrics import roc_auc_score
roc_score = roc_auc_score(dependent, grid.predict_proba(independent)[:,1])
print("The roc_auc score for the model is {}:".format(roc_score))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
cm = confusion_matrix(dependent, y_pred)
clf_report = classification_report(dependent, y_pred)

print(cm)
print(clf_report)
```
