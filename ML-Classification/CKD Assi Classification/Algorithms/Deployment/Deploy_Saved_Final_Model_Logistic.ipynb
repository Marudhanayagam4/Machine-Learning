```python
import pickle 
```

```python
model = pickle.load(open("Final_Logistic_Regression_Model.sav",'rb'))
model
```

```python
scx = pickle.load(open("Standard_Scaler_X.pkl",'rb'))
scx
```

```python
# user inputs

age=int(input("Age: "))
bp=float(input("BP: "))
al=float(input("AL: "))
su=float(input("SU: "))
bgr=float(input("BGR: "))
bu=float(input("BU: "))
sc=float(input("SC: "))
sod=float(input("SOD: "))
pot=float(input("POT: "))
hrmo=float(input("HRMO: "))
pcv=float(input("PCV: "))
wc=float(input("WC: "))
rc=float(input("RC: "))
sg_b=int(input("SG_B 0 or 1 : "))
sg_c=int(input("SG_C 0 or 1 : "))
sg_d=int(input("SG_D 0 or 1 : "))
sg_e=int(input("SG_E 0 or 1 : "))
rbc_normal=int(input("RBC_NORMAL 0 or 1 : "))
pc_normal=int(input("PC_NORMAL 0 or 1 : "))
pcc_present=int(input("PCC_PRESENT 0 or 1 : "))
ba_present=int(input("BA_PRESENT 0 or 1 : "))
htn_yes=int(input("HTN_YES 0 or 1 : "))
dm_yes=int(input("DM_YES 0 or 1 : "))
cad_yes=int(input("CAD_YES 0 or 1 : "))
appet_yes=int(input("APPET_YES 0 or 1 : "))
pe_yes=int(input("PE_YES 0 or 1 : "))
ane_yes=int(input("ANE_YES 0 or 1 : "))

preInput=scx.transform([[age,bp,al,su,bgr,bu,sc,sod,pot,hrmo,pcv,wc,rc,sg_b,sg_c,sg_d,sg_e,rbc_normal,pc_normal,pcc_present,ba_present,htn_yes,dm_yes,cad_yes,appet_yes,pe_yes,ane_yes]])
preInput
```

```python
Future_Predictions = model.predict(preInput)

if Future_Predictions == 0:
    print("NO Chances for CKD Disease")
else:
    print("May have aChances for CKD Disease")
    
print("Chances for CKD Disease YES(1) or NO(0): ", Future_Predictions)
```
